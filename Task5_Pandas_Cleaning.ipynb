```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Cabin'], inplace=True)
```

```python
df['Age'].fillna(df['Age']. median(), inplace=True)
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
rows_before = df.shape[0]
df.drop_duplicates(inplace=True)
rows_after = df.shape[0]

rows_before, rows_after
```

```python
df['Survived'] = df['Survived'].astype('category')
```

```python
df['Pclass'] = df['Pclass'].astype('category')
```

```python
df.info()
```

```python
df['Age_Group'] = pd.cut(
    df['Age'],
    bins=[0, 12, 18, 35, 60, 100],
    labels=['Child', 'Teen', 'Young Adult', 'Adult', 'Senior']
)
```

```python
df['Family_Size'] = df['SibSp'] + df['Parch'] + 1
```

```python
df['Fare_Band'] = pd.qcut(df['Fare'], 4, labels=['Low', 'Medium', 'High', 'Very High'])
```

```python
df.to_csv("cleaned_data.csv", index=False)
```



---## Data Cleaning Notes - Titanic Dataset

1. The Titanic dataset was loaded using pandas `read_csv` for analysis.
2. Initial inspection was done using `head()` and `info()` to understand     structure and datatypes.
3. Missing values were identified using `isnull().sum()`.
4. The Cabin column was dropped due to excessive missing values.
5. Missing Age values were filled using median to reduce the impact of outliers.
6. Missing Embarked values were filled using the mode as it is categorical.
7. Duplicate records were checked and removed to ensure data consistency.
8. Datatypes were converted to categorical where appropriate for better analysis.
9. New features such as Age Group and Family Size were created.
10. The cleaned dataset was exported as a CSV file for further use.
